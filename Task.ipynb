```python

#Convert the Duration of the Songs From Milliseconds to Seconds.
sp_tracks['duration'] = sp_tracks['duration_ms'].apply (lambda x : round(x/1000))
sp_tracks.drop('duration_ms', inplace = True, axis=1)
sp_tracks.duration.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
td = sp_tracks.drop(['key', 'mode', 'explicit'], axis=1).select_dtypes(include=['number'])

# Compute correlation
corr = td.corr(method='pearson')

# Plot heatmap
plt.figure(figsize=(9, 5))
hmap = sns.heatmap(corr, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='Greens')
hmap.set_title('Correlation HeatMap')
hmap.set_xticklabels(hmap.get_xticklabels(), rotation=90)
plt.show()
```

```python
sp_tracks = pd.read_csv('C:/Users/Shubh/Downloads/archive (5)/tracks.csv', index_col='release_date')
sp_tracks.index = pd.to_datetime(sp_tracks.index, errors='coerce')
sp_tracks['dates'] = sp_tracks.index
years = sp_tracks['year'] = sp_tracks['dates'].dt.year
sp_tracks.head()
```

```python
sns.displot(years, discrete=True, aspect=2, height=4, kind='hist',color='g').set(title='No of songs - per year')
```

```python

df = sp_tracks.dropna(subset=['year'])

year_duration = df.groupby('year')['duration_ms'].mean().reset_index()

fig_dims = (15, 5)
fig, ax = plt.subplots(figsize=fig_dims)

sns.barplot(
    data=year_duration,
    x='year',
    y='duration_ms',
    ax=ax,
    errwidth=0
)

ax.set_title('Average Song Duration per Year')
plt.xticks(rotation=90)
plt.show()
```

```python
#spotify feature analysisplt.title('Duration of songs in different Genres')
sns.barplot(y='genre', x='duration_ms', data=sp_feature, hue='genre', palette='crest', dodge=False)
handles, labels = plt.gca().get_legend_handles_labels()
if handles:  
    plt.legend().remove()
    
plt.xlabel('Duration in ms')
plt.ylabel('Genres')
plt.show()
```

```python
#Find top five genres by Popularity and pot a barplot for the same.sns.set_style(style='darkgrid')
plt.figure(figsize=(8,4))
Top = sp_feature.sort_values('popularity', ascending=False)[:10]
sns.barplot(
    y='genre',
    x='popularity',
    data=Top,
    hue='genre',        
    palette='viridis',   
    legend=False       
).set(title='Genres by Popularity-Top 5')
plt.show() 
```
